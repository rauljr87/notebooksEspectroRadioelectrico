### Eliminado las estaciones ESTUDIO
drop() con condición

```python
#  inplace = False, para que no realice los cambios en el mismo DataFrame

estaciones_fm = stn_fm.drop(stn_fm[(stn_fm['TIPO_ESTACION']=='ESTUDIO')].index, inplace = False)
```

```python
estaciones_fm.head()
```

```python
for tipo_estacion, group_tipo_estacion in estaciones_fm.groupby('TIPO_ESTACION'):
    print(tipo_estacion)
```

```python
estaciones_fm.shape
```

```python
estaciones_fm.reset_index(drop=True, inplace=True)
```

```python
estaciones_fm.index
```

## Salvado de datos

#### Método ExcelWriter guarda un libro de excel con varias hojas ####
Se debe de crear un objeto con la clase ExcelWriter de pandas para poder con el realizar el proceso de grabación de varias hojas en un mismo libro de excel

```python
# Creando Objeto ExcelWriter

writer = pd.ExcelWriter('EstacionesEspectra/EstacionesEspectrav3.xlsx')
```

```python
estaciones_fm.to_excel(excel_writer=writer, sheet_name='DataUniverso_FM', index=False, encoding='utf8')
```

```python
estaciones_tv.to_excel(excel_writer=writer, sheet_name='DataUniverso_TV', index=False, encoding='utf8')
```

```python
writer.save()
```

```python
writer.close()
```
